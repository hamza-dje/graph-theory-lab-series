## 1. Libraries
The representation of graphs in Python requires the use of several Well-known python libraries.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Graph Creation Basics
### 2.1. Creation of a Graph
#### 2.1.1. Adding and removing vertices

```python
G = nx.Graph()                      # Undirected graph
G = nx.DiGraph()                    # Directed graph
G.add_node(1)                       # Adds a node labeled by the number 1
G.add_nodes_from([2, 3, 4, 5, 6])   # Adds a list of nodes
G.remove_node(6)                    # Removes the node number 6
G.remove_nodes_from([3, 6])         # Removes a list of nodes
```

#### 2.1.2. Adding and removing edges

```python
G.add_edge(1, 2)                                    # Adds an edge connecting 1 to 2
G.add_edge(1, 4)                                    # Adds an edge connecting 1 to 4
G.add_edges_from([(1, 3), (2, 3), (2, 1), (4, 3)])  # Adds a list of edges
G.remove_edge(1, 3)									# Removes the edge connecting between 1 to 3
G.remove_edges_from([(1, 4), (4, 1)])				# Removes a list of edges
```

### 2.2. Plotting a Graph
#### 2.2.1. Without labels

```python
nx.draw(G)
plt.show()
```

#### 2.2.2. With labels

```python
nx.draw(G, with_labels=True)
plt.show()
```

## 11. Exercise 9 (Graph functions)
### 11.1. Test if a graph is complete

```python
def test_complete(G: nx.DiGraph | nx.Graph) -> bool:
    nodes = list(G.nodes)
    
    if G.is_directed():
        for node in nodes:
            expected_neighbors = [i for i in nodes if i != node]
            if sorted(expected_neighbors) != sorted(list(G[node])):
                return False
        return True
    else:
        for u in nodes:
            for v in [i for i in nodes if i != u]:
                if not G.has_edge(u, v):
                    return False
        return True
```

### 11.2. Test if a graph is reflexive

```python
def test_reflexive(G: nx.DiGraph | nx.Graph) -> bool:
    nodes = list(G.nodes)

    for node in nodes:
        if not G.has_edge(node, node):
            return False
    return True
```

### 11.3. Test if a graph is symmetric

```python
def test_symmetric(G: nx.DiGraph | nx.Graph) -> bool:
    if G.is_directed():
        M = nx.to_numpy_array(G)
        n = len(M)

        for i in range(n):
            for j in range(n):
                if M[i, j] != M[j, i]:
                    return False
        return True
    else:
        return True
```

### 11.4. Test if a graph is transitive

```python
def test_transitive(G: nx.DiGraph| nx.Graph) -> bool:
    nodes = list(G.nodes)

    if G.is_directed():
        for u in nodes:
            for w in list(G[u]):
                for v in list(G[w]):
                    if not v in list(G[u]):
                        return False
        return True
    
    return True
```

### 11.5. Give the complement of a graph

```python
def give_complement(G: nx.Graph) -> nx.Graph:
    if G.is_directed():
        print("No compliment exists for a directed graph")

    nodes = list(G.nodes)
    n = len(nodes)
    complement = nx.Graph()
    complement.add_nodes_from(nodes)

    for i in range(n):
        for j in range(i + 1, n):
            u, v = nodes[i], nodes[j]
            if not G.has_edge(u, v):
                complement.add_edge(u, v)
            
    return complement
```

### 11.6. Generate subgraph

```python
def generate_subgraph(G: nx.Graph | nx.DiGraph, subset: list[any]) -> nx.Graph | nx.DiGraph:
    H = G.__class__()

    for node in subset:
        if node in G.nodes:
            H.add_node(node)

    for (u, v) in G.edges:
        if u in H.nodes and v in H.nodes:
            H.add_edge(u, v)

    return H
```
